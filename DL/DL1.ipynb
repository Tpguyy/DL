```python
import numpy as np

class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=10):
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand()
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activate_bipolar(self, x):
        return 1 if x > 0 else -1

    def activate_unipolar(self, x):
        return 1 if x > 0 else 0

    def train(self, inputs, labels):
        print("Epoch | Input x1 | Input x2 | Desired Output | Net Input Bias | Error | Weights | Change in Weights")
        
        for epoch in range(self.epochs):
            for i in range(len(inputs)):
                prediction = self.predict(inputs[i])
                error = labels[i] - prediction
                delta_weights = self.learning_rate * error * inputs[i]
                self.weights += delta_weights
                self.bias += self.learning_rate * error
                print(f"{epoch+1:5} | {inputs[i][0]:9} | {inputs[i][1]:9} | {labels[i]:15} | {np.dot(inputs[i], self.weights) + self.bias:17.2}")

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return self.activate_unipolar(summation)

# AND gate
and_gate_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
and_gate_labels = np.array([0, 0, 0, 1])
and_gate_perceptron = Perceptron(input_size=2)
and_gate_perceptron.train(and_gate_inputs, and_gate_labels)
```
